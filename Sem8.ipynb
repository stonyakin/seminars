# Субдифференциал

## 1. Субдифференциал. Субградиент.

Вектор $g(x)$ называется **субградиентом** функции $f$ в точке $x$, если $\forall y\in X$ выполнено

$$f(y) \geq f(x) + \langle g(x), y - x \rangle$$

**Субдифференциал** $\partial f(x)$ - это множество всех субградиентов.

Пример

$\partial \|x\|_2=?$

$$\nabla \|x\|_2 = \frac{x}{\|x\|}, x\neq 0 $$

$$\partial \|x\| = \left\{\frac{x}{\|x\|}\right\}, x\neq 0 $$

$x = 0$:

$$\|y\| \geq \langle g, y\rangle, \forall y \in \mathbb{R}^n$$

$$\partial f(0) = \{g | \|y\| \geq \langle g, y\rangle, \forall y \in \mathbb{R}^n\}$$

$B_1(0) \subseteq \partial f(0),$ поскольку $\langle g, y\rangle \leq \|g\| \|y\|\leq \|y\|$.

Допустим, $g\in\partial f(0): \|g\|>1$, то для $y=g$ $\langle g, y\rangle= \|g\|^2 > \|g\| = \|y\|$.

$$\partial f(0)= B_1(0)$$

## 2. Свойства Субдифференциала

1. Если функция дифференцируемая и выпукла, то $\partial f(x) = \{\nabla f(x)\}$.

2. $\partial f(\alpha x) = \alpha \partial f(y)|_{y=\alpha x}$

3. $\partial f(A x + b) = A^\top \partial f(y)|_{y=Ax+b}$

4. (Теорема Моро-Ракофеллара) Пусть $f(x)=\sum\limits_{i=1}^n a_i f_i(x)$, $f_i$-выпуклые на $X$ функции, $a_i\geq 0$. Тогда на множестве $X$ функция $f$ имеет субдифференциал $\partial f(x) = \sum\limits_{i=1}^n a_i \partial f_i(x)$

5. Если $f(x)=\max\limits_{i=\overline{1,k}}f_i(x)$, то $$\partial f(x)=Conv\left(\bigcup\limits_{i\in I(x)}\partial f_i(x)\right),$$
где $I(x) = \{i|f_i(x)=\max f(x)\}$.

$Conv(X) = \{\alpha x_1 +(1-\alpha)x_2| \forall \alpha \in[0,1], x_1, x_2 \in X\}$

$f(x)= \|Ax-b\|_1 = \sum\limits_{i=1}^m |a_i^\top x - b_i|$

$\partial \|Ax-b\|_1 = A^\top \partial \|y\|_1\Big|_{y=Ax-b} = A^\top \text{sign}(Ax-b)$

$$\partial \|y\|_1 = \partial \sum_{i=1}^n |y_i|=\sum_{i=1}^n \partial |y_i| =\sum_{i=1}^n  e_i \text{sign}(y_i) = \text{sign}(y)$$

## 3. Примеры

1. $f(x) = \max(e^x, 1-x)$

  $x<0:\partial f(x) = \partial (1-x) = -1$

  $x>0:\partial f(x) = \partial (e^x) = e^x$

  $x=0: \partial f(0)= Conv(\partial (1-x)|_{x=0} \cup \partial e^x|_{x=0})= Conv(\{-1, 1\})=[-1,1]$


2. $f(x) = |x| = \max\left(-x, x\right)$

  $x<0:\partial f(x) = \partial (-x) = -1$

  $x>0:\partial f(x) = \partial (x) = 1$

  $x=0: \partial f(0)= Conv(\partial (-x)|_{x=0} \cup \partial x|_{x=0})= Conv(\{-1, 1\})=[-1,1]$

3. $f(x)=|c^\top x| = \max(-c^\top x, c^\top x), x\in\mathbb{R}^n$

  $c^\top x < 0: \partial f(x) = \partial (-c^\top x)=-c$

  $c^\top x > 0: \partial f(x) = \partial (c^\top x)=c$

  $c^\top x = 0: \partial f(x) = Conv(\partial (c^\top x)\cup \partial (-c^\top x) ) = Conv(\{-c, c\})=\{-c\alpha +(1-\alpha) c| \alpha \in [0,1]\}=[-c, c]$

  $\partial |c^\top x| = c \partial |y|\Big|_{y=c^\top x}$

4. $f(x)=|c_1^\top x| + |c_2^\top x|$

  $\partial f(x) = \partial |c_1^\top x| + \partial |c_2^\top x|$

  $c_1^\top x \neq 0, c_2^\top x \neq 0: \partial f(x)= sign(c_1^\top x) c_1 + sign(c_2^\top x)c_2$

  $c_1^\top x = 0, c_2^\top x \neq 0: \partial f(x)= [-c_1, c_1] + sign(c_2^\top x)c_2$

  $c_1^\top x \neq  0, c_2^\top x = 0: \partial f(x)= sign(c_1^\top x)c_1 + [-c_2, c_2]$

  $c_1^\top x =  0, c_2^\top x = 0: \partial f(x)= \{-\alpha_1 c_1 + \alpha_2 c_1 - \alpha_3 c_2 +\alpha_4 c_2|\sum\limits_{i=1}^4 \alpha_i = 1, \alpha_i\geq 0\}$
